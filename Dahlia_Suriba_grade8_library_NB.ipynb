```python
unique_book_ids = set()
for key, book_info in books_data.items():
    # Assuming each dictionary in books_data represents a book and has an 'id'
    if isinstance(book_info, dict) and 'id' in book_info:
        unique_book_ids.add(book_info['id'])

total_unique_books = len(unique_book_ids)
print(f"Total number of books: {total_unique_books}")
```
